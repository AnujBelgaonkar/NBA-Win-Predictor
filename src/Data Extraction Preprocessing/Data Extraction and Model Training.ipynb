```python
from nba_api.stats.static import teams
import pandas as pd
import numpy as np
import time
```

```python
temp = teams.get_teams()
```

```python
temp
```

```python
team_ids = []
```

```python
for team in temp:
    team_ids.append([team['abbreviation'],team['id']])
```

```python
team_ids
```

```python
seasons = ['2021-22','2022-23','2023-24']
```

```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable = 1610612737,
                                            season_nullable = seasons)
```

```python
games = gamefinder.get_data_frames()[0]
```

```python
print(games)
```

```python
def get_team_games(seasons,teams):
    temp = pd.DataFrame()
    i = 0
    for team in teams:
        i += 1
        for season in seasons:
            gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable = team,
                                                          season_nullable = season,
                                                    season_type_nullable = 'Regular Season')
            temp = pd.concat([temp,gamefinder.get_data_frames()[0]])
        if(i == 0):
            time.sleep(5)
    return temp
```

```python
start = time.time()
temp_data = get_team_games(seasons,team_ids)
end = time.time()
```
